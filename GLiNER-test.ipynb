```python
from gliner import GLiNER
```

```python
model = GLiNER.from_pretrained("urchade/gliner_base")
```

Taking a page out of a GSWA WMC - BHP Billtion Partial Surrender Report from the late 90s

https://geodocs.dmirs.wa.gov.au/Web/documentlist/10/Report_Ref/A55767

```python
text = """In 1995 a regional gravity survey was conducted by WMC over its leases in the Kambalda
area. The data was collected on a 1km x 1km grid. The data from this survey, covering the
relinquished portion of the lease is located on the data tape in Appendix I.
A surface transient electromagnetic (TEM) survey was conducted in 1995 over the
perceived prospective mafic/ultramafic contacts in the tenement area. A small portion of
these surveys extended onto the relinquished portion of the tenement. No significant
anomalies were identified over the relinquished portion of the tenement.
A total of 13 aircore holes were drilled by WMC on the relinquished ground (Figure 12). The
results were generally disappointing with very low to background nickel and gold values.
Results are presented in Appendix Il.
Exploration on E15/101 during this period was reported by Olsen and Smillie (1994), Olsen
(1995, 1996), and Offe et al (1996).
1997 - 1998 Acacia Resources
When Acacia Resources gained exploration access to the western portion of E15/101 they
decided that the project area should be completely resampled using a consistent sampling
technique, as the variability of the Billiton sampling and the broad spacing of the
composites, had cast some doubt on the reliability of the gold anomalies generated by this
work. The entire Black Knob project area was therefore resampled on a 100 x 200 metre
staggered grid. A total of 745 samples was taken.
Samples were collected using an auger rig mounted on a 4WD, and using a DGPS to locate
the sample sites. The depth of sample collection was no less than 30 centimetres and
samples were taken from the most calcareous layer. The reaction of the sample with HCl
acid was used to determine the relative amount of carbonate within the soil. Samples of
approximately one kilogram were collected and submitted to Genalysis laboratories in Perth
for analysis for gold (1ppb detection limit) by aqua regia digest with an AAS graphite furnace
finish and arsenic (5ppm) and nickel ({ppm) by aqua regia digest with a standard AAS
finish. Standards, blanks and duplicates were submitted as quality control. Soil sampling
logs and assays are presented in the PC format floppy disk in Appendix Ill. The data format
is space delimited ASCII files.
Probability statistics were used to determine anomalous gold, nickel, and arsenic
populations from the assay results. The results on the whole were largely disappointing
however a number of moderately anomalous zones of gold, nickel, and arsenic
mineralisation were delineated.
A broad (2000 x 1900m), moderately anomalous (maximum 29ppb) zone of gold
mineralisation was identified in the northern portion of the project area and several smaller
clusters were delineated throughout the southern portion. A number of clusters of elevated
nickel (maximum 620ppm) coincident with anomalous arsenic (maximum 115ppm) are
present, although at least one of these anomalies is associated with drainage.
Steuart (1998) summarised the work on the western portion of tenement E15/101 for the
period January 1997 to January 1998.
EXPENDITURE
Total expenditure by WMC on the surrendered ground from January 1989 to January 1996

was $19,720. A breakdown of the known costs is presented in the table below: 
"""
```

```python
labels = ["geophysics", "licence", "commodity", "date", "company", "project", "survey"]

entities = model.predict_entities(text, labels, threshold=0.5)

for entity in entities:
    print(entity["text"], "=>", entity["label"])
    
## results from a script version

1995 => date
regional gravity survey => survey
WMC => company
surface transient electromagnetic => geophysics
1995 => date
WMC => company
nickel => commodity
gold => commodity
E15/101 => project
1995 => date
1996), and Offe et al (1996 => date
1997 - 1998 => date
Acacia Resources => company
Acacia Resources => company
E15/101 => project
Billiton => company
gold => commodity
Black Knob => project
DGPS => geophysics
Genalysis laboratories => company
gold => commodity
arsenic => commodity
nickel => commodity    
```
